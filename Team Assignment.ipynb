```python
import pandas as pd # data science essentials
import matplotlib.pyplot as plt # data visualization
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm


plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
# specifying a file (must be in the working directory)
file = 'Working dataset - Python.xlsx'

# reading the file into Python through pandas
middleast = pd.read_excel(file)

middleast.describe().round(2)
data=middleast

df_dropped = pd.DataFrame.copy(middleast)
df_dropped = df_dropped.dropna().round(2)


plt.rcParams['font.size']=11 
plt.rcParams['figure.figsize']=[8.0, 3.5]

 

fig = plt.figure(figsize = (10, 8))
axes = fig.subplots(nrows=2, ncols=1)

middleast.plot(kind='scatter', x='GDP (current US$)', y='Death rate, crude (per 1,000 people)', ax=axes[0], color='Blue'); 

middleast.plot(kind='scatter', x='GDP (current US$)', y='Birth rate, crude (per 1,000 people)', ax=axes[1], color='Red'); 

ax = sns.regplot(x='GDP (current US$)', y='Birth rate, crude (per 1,000 people)', data=middleast)

palette = sns.color_palette("Paired", 10)
sns.set_palette(palette)

print(middleast[['GDP (current US$)','Country Name']])


plt.tight_layout()

```
